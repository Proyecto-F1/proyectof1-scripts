```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_2021 = pd.read_csv('2021_overtakes.csv', delimiter=(','))
df_2021['Año'] = 2021
df_2021.head()
```

```python
frequencies =(df_2021['Race'].value_counts())

fig, ax = plt.subplots(figsize=(10, 25))
ax.barh(frequencies.index, frequencies.values)
ax.set_title("Adelantamientos por circuito")
ax.set_ylabel("Circuito")
ax.grid(axis = 'x')
ax.set_xlabel('Número de adelantamientos')
```

```python
frequencies =(df_2021['Lap'].value_counts())

fig, ax = plt.subplots(figsize=(10, 25))
ax.barh(frequencies.index, frequencies.values)
ax.set_title("Adelantamientos por vuelta")
ax.set_ylabel("Vuelta")
ax.grid(axis = 'x')
ax.set_xlabel('Número de adelantamientos')
```

```python
frequencies =(df_2021['Overtaken'].value_counts())

fig, ax = plt.subplots(figsize=(10, 25))
ax.barh(frequencies.index, frequencies.values)
ax.set_title("Piloto más adelantado")
ax.set_ylabel("Piloto")
ax.grid(axis = 'x')
ax.set_xlabel('Número de veces que le han adelantado')
```

```python
frequencies =(df_2021['Overtaker'].value_counts())

fig, ax = plt.subplots(figsize=(10, 25))
ax.barh(frequencies.index, frequencies.values)
ax.set_title("Piloto que más adelanta")
ax.set_ylabel("Piloto")
ax.grid(axis = 'x')
ax.set_xlabel('Número de adelantamientos')
```

```python
frequencies =(df_2021['Position'].value_counts())

fig, ax = plt.subplots(figsize=(10, 25))
ax.barh(frequencies.index, frequencies.values)
ax.set_title("Posición a la que pasa al realizar el adelantamiento")
ax.set_ylabel("Posición")
ax.grid(axis = 'x')
ax.set_xlabel('Número de adelantamientos en dicha posición')
```

```python
lista = []
carreras = df_2021['Race']
for elemento in carreras:
    if elemento not in lista:
        lista.append(elemento)
        
for elemento in lista:
    frequencies = df_2021[df_2021['Race'] == elemento]
    

    h =(frequencies['Overtaker'].value_counts())

    fig, ax = plt.subplots(figsize=(len(h.index)*1.2, len(h.values)))
    ax.barh(h.index, h.values)
    ax.set_xlabel('Adelantamientos realizados')
    ax.set_ylabel("Piloto")
    ax.grid(axis = 'x')
    ax.set_title(elemento)


```
