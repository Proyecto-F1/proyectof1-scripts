```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df_2014 = pd.read_csv('2014_overtakes.csv', delimiter=(','))
df_2014['Año'] = 2014
df_2014.head()

df_2015 = pd.read_csv('2015_overtakes.csv', delimiter=(','))
df_2015['Año'] = 2015
df_2015.head()

df_2016 = pd.read_csv('2016_overtakes.csv', delimiter=(','))
df_2016['Año'] = 2016
df_2016.head()

df_2017 = pd.read_csv('2017_overtakes.csv', delimiter=(','))
df_2017['Año'] = 2017
df_2017.head()

df_2018 = pd.read_csv('2018_overtakes.csv', delimiter=(','))
df_2018['Año'] = 2018
df_2018.head()

df_2019 = pd.read_csv('2019_overtakes.csv', delimiter=(','))
df_2019['Año'] = 2019
df_2019.head()

df_2020 = pd.read_csv('2020_overtakes.csv', delimiter=(','))
df_2020['Año'] = 2020
df_2020.head()

df_2021 = pd.read_csv('2021_overtakes.csv', delimiter=(','))
df_2021['Año'] = 2021
df_2021.head()

df_2022 = pd.read_csv('2022_overtakes.csv', delimiter=(','))
df_2022['Año'] = 2022
df_2022.head()



df = df_2014
df = df.append(df_2015)
df = df.append(df_2016)
df = df.append(df_2017)
df = df.append(df_2018)
df = df.append(df_2019)
df = df.append(df_2020)
df = df.append(df_2021)
df = df.append(df_2022)

df.head()
```

```python
df_race = df.groupby(by=["Race", "Año"]).count().apply(list)

df_race = df_race["Lap"].to_frame()

result = {}

for index, row in df_race.iterrows():
    if index[0] in result:
        result[index[0]].append(row["Lap"])
    else:
        result[index[0]] = [row["Lap"]]
        
result_list = []
xticklabels = []
data = result
# build a box plot

for key in result:
    data_list = list(data[key])
    if len(data_list) > 0:
        result_list.append(data_list)
        xticklabels.append(key)
        
fig, ax = plt.subplots(figsize=(len(xticklabels)*1.2, len(xticklabels)))

ax.boxplot(result_list)

# title and axis labels
ax.set_title('box plot')
ax.set_xlabel('circuitos')
ax.set_ylabel('adelantamientos')
ax.set_xticklabels(xticklabels, rotation = 45 , ha="right")
```

```python
dic = {}
lista1 = []

# eliminamos las carreras con menos de tres particiones entre 2014 y 2022
# ya que no podemos observar la distribución con tan pocos datos

for k,v in data.items():
    if len(v) < 3:
         lista1.append(k)
      
# sumamos los adelantamientos totales por circuito durante los años 2014-2022

for elemento in lista1:
    del data[elemento]
      
for k,v in data.items():
    sum = 0
    for elemento in v:
        sum += elemento
    
    dic[k] = sum / len(v) # dividimos entre la media, ya que no todos los circuitos se ha participado los mismos años
dic
```

```python
lista = []

# si la carrera es un Sprint la eliminamos
# ya que es insignificante para el código predictivo que vamos a desarrollar (objetivo final)

for elemento in dic.keys():
    lista.append(elemento)
    
for elemento in lista:
    if 'Sprint' in elemento:
        del dic[elemento]
dic
```

```python
# ordenamos los circuitos por el total de adelantamientos (de menos a más)

import operator
ordenados = sorted(dic.items(), key=operator.itemgetter(1))
ordenados
```

```python
# Cogemos los 5 valores más grandes, y los 5 más pequeños
# ya que son los más importantes para la distribución

lista_final = ordenados[:5] + ordenados[-5:]
lista_final

# Medias de adelantamientos por circuito (5 mayores y 5 menores)
```

```python
# Creamos de nuevo un diccionario con los valores de adelantamientos por año
# para hacer el boxplot (distribución)

l = []
d = {}
for elemento in lista_final:
    l.append(elemento[0])

    
for elemento in l:
    d[elemento] = data[elemento]
    
    
    
d
```

```python
# RESULTADO FINAL DISTRIBUCIÓN

result_list = []
xticklabels = []
data = d


# build a box plot

for key in d:
    data_list = list(data[key])
    if len(data_list) > 0:
        result_list.append(data_list)
        xticklabels.append(key)
        
fig, ax = plt.subplots(figsize=(18,18))
ax.boxplot(result_list)








# title and axis labels
ax.set_title('box plot')
ax.set_xlabel('circuitos')
ax.set_ylabel('adelantamientos')
ax.set_xticklabels(xticklabels, ha = "right")
plt.tight_layout()

# Guardar el gráfico en un archivo
plt.savefig('boxplot.png', dpi=300)
```
