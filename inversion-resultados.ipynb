```python
import pandas as pd
import numpy as np
df_inversion=pd.read_csv('Inversión de los equipos F1 en 2021  - Hoja 1.csv',delimiter=',')
df_eficiencia_pits=pd.read_csv('Eficiencia en podios - Hoja 1.csv',delimiter=',')
df_empleados=pd.read_csv('Numero empleados - Hoja 1.csv',delimiter=',')
df_podios=pd.read_csv('Podios - Hoja 1.csv',delimiter=',')
df_empleados = df_empleados.rename(columns={"Escudería": "Esqudería"})
df_inversion =df_inversion.rename(columns={"Escudería": "Esqudería"})
df_inversion['Presupuesto total']=df_inversion['Presupuesto total'].map(lambda x: x.replace(' millones','').replace('$', ''))

```

```python
import matplotlib.pyplot as plt

# Creamos el gráfico de dispersión
fig, ax = plt.subplots(figsize=(8, 8))
plt.scatter(df_inversion['Presupuesto total'], df_inversion['Puntos'])

# Añadimos las etiquetas de los ejes
plt.xlabel('Presupuesto total')
plt.ylabel('Puntos')

# Agregamos el nombre de cada escudería a cada punto
for i, escuderia in enumerate(df_inversion['Esqudería']):
    plt.annotate(escuderia, (df_inversion['Presupuesto total'][i], df_inversion['Puntos'][i]))
plt.title('Relación entre presupuesto y puntos año 2021 ')
# Mostramos el gráfico
plt.show()


df_inversion['Presupuesto total'] = df_inversion['Presupuesto total'].str.replace('$', '').str.replace('millones', '').astype(float)

correlation = df_inversion['Presupuesto total'].corr(df_inversion['Puntos'], method='pearson')
print('Correlación entre presupuesto total y puntos: ', correlation)
plt.savefig('Relacion presupuesto-puntos.png')
```
